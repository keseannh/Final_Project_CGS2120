```python
import math
from ipywidgets import widgets, VBox

# Arithmetic Functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero."
    return a / b

def power(a, b):
    return a ** b

def square_root(a):
    if a < 0:
        return "Error: Square root of a negative number."
    return math.sqrt(a)

# Function to perform calculation
def calculate(operation, num1, num2):
    """Performs the selected operation."""
    try:
        if operation == "Addition":
            return f"Result: {add(num1, num2)}"
        elif operation == "Subtraction":
            return f"Result: {subtract(num1, num2)}"
        elif operation == "Multiplication":
            return f"Result: {multiply(num1, num2)}"
        elif operation == "Division":
            return f"Result: {divide(num1, num2)}"
        elif operation == "Exponentiation":
            return f"Result: {power(num1, num2)}"
        elif operation == "Square Root":
            return f"Result: {square_root(num1)}"
        else:
            return "Invalid operation."
    except Exception as e:
        return f"Error: {e}"

# Callback for button click
def on_calculate_button_click(change):
    """Updates the result when the calculate button is clicked."""
    operation = operation_dropdown.value
    num1 = num1_input.value
    num2 = num2_input.value if operation != "Square Root" else 0
    result_label.value = calculate(operation, num1, num2)

# Widgets
operation_dropdown = widgets.Dropdown(
    options=["Addition", "Subtraction", "Multiplication", "Division", "Exponentiation", "Square Root"],
    description="Operation:",
    style={'description_width': 'initial'}
)

num1_input = widgets.FloatText(
    value=0.0,
    description="First Number:",
    style={'description_width': 'initial'}
)

num2_input = widgets.FloatText(
    value=0.0,
    description="Second Number:",
    style={'description_width': 'initial'}
)

calculate_button = widgets.Button(
    description="Calculate",
    button_style='success',
    tooltip="Click to perform the calculation",
    icon='check'
)

result_label = widgets.Label(value="Result: ")

# Link button click to calculation
calculate_button.on_click(on_calculate_button_click)

# Layout and display
display(VBox([operation_dropdown, num1_input, num2_input, calculate_button, result_label]))
```
